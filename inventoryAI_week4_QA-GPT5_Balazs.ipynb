````python
# -*- coding: utf-8 -*-
# INVENTORYAI - 4. h√©t: K√©rd√©s-v√°lasz modell kipr√≥b√°l√°sa t√∂bb modellel
# Notebook neve: inventoryAI_week4_QA-multimodel.ipynb

# -------------------------------
# 1Ô∏è‚É£ K√∂rnyezet √©s csomagok import√°l√°sa
# -------------------------------
import os
from dotenv import load_dotenv
from openai import OpenAI

# -------------------------------
# 2Ô∏è‚É£ .env bet√∂lt√©se √©s OpenAI kulcs ellen≈ërz√©se
# -------------------------------
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise RuntimeError("‚ùå Nincs be√°ll√≠tva az OPENAI_API_KEY k√∂rnyezeti v√°ltoz√≥!")

client = OpenAI(api_key=api_key)

# -------------------------------
# 3Ô∏è‚É£ Felhaszn√°l√≥i k√©rd√©s bek√©r√©se
# -------------------------------
user_question = input("√çrj be egy k√©rd√©st (pl. Mi az a mesters√©ges intelligencia?): ")
print(user_question)

# -------------------------------
# 4Ô∏è‚É£ Kipr√≥b√°land√≥ modellek list√°ja
# -------------------------------
models_to_test = [
    "gpt-4o-mini",
    "gpt-4-turbo",
    "gpt-5"
]

# -------------------------------
# 5Ô∏è‚É£ K√©rd√©s-v√°lasz gener√°l√°sa t√∂bb modellel
# -------------------------------
for model_name in models_to_test:
    print(f"\n==============================")
    print(f"ü§ñ Modell: {model_name}")
    print("==============================")

    try:
        response = client.chat.completions.create(
            model=model_name,
            messages=[
                {"role": "system", "content": "Te egy seg√≠t≈ëk√©sz egyetemi asszisztens vagy."},
                {"role": "user", "content": user_question}
            ]
        )

        answer = response.choices[0].message.content
        print(f"\nüìù V√°lasz:\n{answer}\n")

    except Exception as e:
        print(f"‚ö†Ô∏è Hiba t√∂rt√©nt a(z) {model_name} modelln√©l: {e}")

# -------------------------------
# 6Ô∏è‚É£ √ñsszegz√©s
# -------------------------------
print("\n‚úÖ √ñsszehasonl√≠t√°s k√©sz!")
print("A fenti modellek v√°laszai alapj√°n eld√∂ntheted, melyik ad term√©szetesebb vagy pontosabb v√°laszt.")
````
